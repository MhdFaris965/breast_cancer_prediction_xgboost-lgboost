```python
X=df.drop(['diagnosis'],axis=1)
y=df['diagnosis']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
import xgboost as xgb
import lightgbm as lgb
```

```python
xgb_model=xgb.XGBClassifier()
lgb_model=lgb.LGBMClassifier()
```

```python
xgb_model.fit(X_train,y_train)
lgb_model.fit(X_train,y_train)
```

```python
xgb_y_pred_train=xgb_model.predict(X_train)
lgb_y_pred_train=lgb_model.predict(X_train)

xgb_y_pred_test=xgb_model.predict(X_test)
lgb_y_pred_test=lgb_model.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print('xgb_traing')
print('-'*50)
print(classification_report(y_train,xgb_y_pred_train))

print('xgb_testing')
print('-'*50)
print(classification_report(y_test,xgb_y_pred_test))

print('lgb_traing')
print('-'*50)
print(classification_report(y_train,lgb_y_pred_train))

print('lgb_testing')
print('-'*50)
print(classification_report(y_test,lgb_y_pred_test))
```
